```python
#print dataframe eg
import pandas as pd

print("pandas dataframe example")

data_dict = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 30, 35, 40, 45],    
        'marks': [85, 90, 95, 80, 75]}

df = pd.DataFrame(data_dict)
print(df)
print("dataframe:\n", df)
print(df.iloc[1])
print(df.iloc[0:1])
print(df.iloc[0:3])
print(df['Age'])
print(df.head())
```
